```python
import ctypes
import random as rd
from numpy import argsort
```

```python
bucket = ctypes.CDLL('./BucketSort.so')
bucket.bucket_sort.restype = ctypes.POINTER(ctypes.c_uint64)
bucket.bucket_sort.argtypes = [ctypes.POINTER(ctypes.c_uint64), ctypes.c_uint64, ]
```

```python
radix = ctypes.CDLL('../RadixSort/RadixSort.so')
radix.radix_sort.restype = ctypes.POINTER(ctypes.c_uint64)
radix.radix_sort.argtypes = [ctypes.POINTER(ctypes.c_uint64), ctypes.c_uint64, ]
```

```python
pyarr = [rd.randint(0,9) for i in range(1000)]
```

```python
pyarr2 = []
count = [0 for i in range(10)]
for i in pyarr:
    pyarr2.append(count[i])
    count[i]+=1
```

```python
arr = (ctypes.c_uint64 * len(pyarr))(*pyarr)
```

```python
res = bucket.bucket_sort(ctypes.cast(ctypes.byref(arr), ctypes.POINTER(ctypes.c_uint64)), len(pyarr))
pyres = [res[i] for i in range(len(pyarr))]
```

```python
r_res = radix.radix_sort(ctypes.cast(ctypes.byref(arr), ctypes.POINTER(ctypes.c_uint64)), len(pyarr))
r_pyres = [r_res[i] for i in range(len(pyarr))]
```
